```python
import numpy as np
import pandas as pd
import scipy
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
pop1 = np.random.binomial(10, 0.2, 10000)
pop2 = np.random.binomial(10,0.5, 10000) 

sample1 = np.random.choice(pop1, 100, replace=True)
sample2 = np.random.choice(pop2, 100, replace=True) 

plt.hist(sample1, alpha=0.5, label='sample 1') 
plt.hist(sample2, alpha=0.5, label='sample 2')
plt.legend()

print(sample1.mean())
print(sample2.mean())
print(sample1.std())
print(sample2.std())

```

```python
#1. Sample changed to 1000
sample1a = np.random.choice(pop1, 1000, replace=True)
sample2a = np.random.choice(pop2, 1000, replace=True) 

plt.hist(sample1a, alpha=0.5, label='sample 1a') 
plt.hist(sample2a, alpha=0.5, label='sample 2a')
plt.legend()

print(sample1a.mean())
print(sample2a.mean())
print(sample1a.std())
print(sample2a.std())
print("Both the means and standard deviations change")
```

```python
#1. Sample changed to 20
sample1b = np.random.choice(pop1, 1000, replace=True)
sample2b = np.random.choice(pop2, 1000, replace=True) 

plt.hist(sample1b, alpha=0.5, label='sample 1b') 
plt.hist(sample2b, alpha=0.5, label='sample 2b')
plt.legend()

print(sample1b.mean())
print(sample2b.mean())
print(sample1b.std())
print(sample2b.std())
print("Both the means and standard deviations change")
```

```python
#2. Change probability value
pop1a = np.random.binomial(10, 0.3, 10000)
pop2a = np.random.binomial(10,0.5, 10000) 

sample2a = np.random.choice(pop1a, 100, replace=True)
sample3a = np.random.choice(pop2a, 100, replace=True) 

plt.hist(sample2a, alpha=0.5, label='sample 2a') 
plt.hist(sample3a, alpha=0.5, label='sample 3a')
plt.legend()

print(sample2a.mean())
print(sample3a.mean())
print(sample2a.std())
print(sample3a.std())

diff=sample3a.mean() - sample2a.mean()
print("mean difference:",diff)

sd = np.array([sample2a.std(), sample3a.std()])
diff_se = (sum(sd**2 / 100)) ** .5
print("t-value:", diff/diff_se)

from scipy.stats import ttest_ind
print(ttest_ind(sample3a, sample2a, equal_var=False))
```

```python
#2.p value for group one changed to 0.4
pop1b = np.random.binomial(10, 0.4, 10000)
pop2b = np.random.binomial(10,0.5, 10000) 

sample4b = np.random.choice(pop1a, 100, replace=True)
sample5b = np.random.choice(pop2a, 100, replace=True) 

plt.hist(sample2a, alpha=0.5, label='sample 4b') 
plt.hist(sample3a, alpha=0.5, label='sample 5b')
plt.legend()

print(sample4b.mean())
print(sample5b.mean())
print(sample4b.std())
print(sample5b.std())

diff=sample5b.mean() - sample4b.mean()
print("mean difference:",diff)

sd = np.array([sample4b.std(), sample5b.std()])
diff_se = (sum(sd**2 / 100)) ** .5
print("t-value:", diff/diff_se)

from scipy.stats import ttest_ind
print(ttest_ind(sample5b, sample4b, equal_var=False))
print("The samples get more similar and the p value decreases")
```
