```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline 
```

```python
#For each distribution....
```

```python
Exponential = np.random.exponential(scale=1, size=100)
mean = np.mean(Exponential)
std = np.std(Exponential, ddof=1)
print("Mean:", mean)
print("Std:", std)
plt.hist(Exponential)
plt.axvline(mean, color='b', linestyle='solid', linewidth=2)
plt.axvline(mean + std, color='b', linestyle='dashed', linewidth=2) 
plt.axvline(mean - std, color='b', linestyle='dashed', linewidth=2) 
plt.show()
#This is a skewed distribution. The mean is not in the center and can not be a 
#measure of central tendency. The standandard deviation doesn't encompass the 
#same amount of data in either side and hence doesn't measure the varience between values

```

```python
Uniform = np.random.uniform(low=2000, high=6000, size=100)
plt.hist(Uniform)
mean = np.mean(Uniform)
std = np.std(Uniform, ddof=1)
print("Mean:", mean)
print("Std:", std)
plt.axvline(mean, color='b', linestyle='solid', linewidth=2)
plt.axvline(mean + std, color='b', linestyle='dashed', linewidth=2) 
plt.axvline(mean - std, color='b', linestyle='dashed', linewidth=2) 
plt.show()
plt.show()
#The mean is at the center and the standard deviation represents equal amount of data in either dide.
```

```python
Triangular = np.random.triangular(18, 26, 35, 100)
mean = np.mean(Triangular)
std = np.std(Triangular, ddof=1)
print("Mean:", mean)
print("Std:", std)
plt.hist(Triangular)
plt.axvline(mean, color='b', linestyle='solid', linewidth=2)
plt.axvline(mean + std, color='b', linestyle='dashed', linewidth=2) 
plt.axvline(mean - std, color='b', linestyle='dashed', linewidth=2) 
plt.show()
#The mean is at the center and the standard deviation represents equal amount of data in either dide.
```

```python
Power = np.random.power(1, 100)
mean = np.mean(Power)
std = np.std(Power, ddof=1)
print("Mean:", mean)
print("Std:", std)
plt.hist(Power)
plt.axvline(mean, color='b', linestyle='solid', linewidth=2)
plt.axvline(mean + std, color='b', linestyle='dashed', linewidth=2) 
plt.axvline(mean - std, color='b', linestyle='dashed', linewidth=2) 
plt.show()
#The distribution is skewed but the looks to be in the middle
```

```python
Geometric = np.random.geometric(.50, 100)
mean = np.mean(Geometric)
std = np.std(Geometric, ddof=1)
print("Mean:", mean)
print("Std:", std)
plt.hist(Geometric)
plt.axvline(mean, color='b', linestyle='solid', linewidth=2)
plt.axvline(mean + std, color='b', linestyle='dashed', linewidth=2) 
plt.axvline(mean - std, color='b', linestyle='dashed', linewidth=2) 
plt.show()
#The mean is not a measure of central tendency and the standard deviation doesn't
#represent the varience within the data.
```

```python
Gumbel = np.random.gumbel(0, 1, 100)
mean = np.mean(Gumbel)
std = np.std(Gumbel, ddof=1)
print("Mean:", mean)
print("Std:", std)
plt.hist(Gumbel)
plt.axvline(mean, color='b', linestyle='solid', linewidth=2)
plt.axvline(mean + std, color='b', linestyle='dashed', linewidth=2) 
plt.axvline(mean - std, color='b', linestyle='dashed', linewidth=2) 
plt.show()
```

```python
#The mean is generally at the center and not equal data is covered in either side of
#the standard deviation
```

```python
Var1 = np.random.normal(5, 0.5, 100)
Var2 = np.random.normal(10, 1, 100)
Var = Var1 + Var2
mean = np.mean(Var)
std = np.std(Var, ddof=1)
print("Mean:", mean)
print("Std:", std)
plt.hist(Var)
plt.axvline(mean, color='b', linestyle='solid', linewidth=2)
plt.axvline(mean + std, color='b', linestyle='dashed', linewidth=2) 
plt.axvline(mean - std, color='b', linestyle='dashed', linewidth=2) 
plt.show()
#The distribution doesn't seem symetrical anymore.
```
