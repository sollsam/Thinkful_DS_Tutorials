```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline 
```

```python
df = pd.read_csv('un.csv')
df.head()
```

```python
df.dropna(inplace=True)
df.head()
plt.hist(df['lifeMale']) 

plt.show()
```

```python
plt.boxplot(df['lifeMale'])
plt.show()
```

```python
plt.figure(figsize=(10, 5))

plt.subplot(1, 2, 1)
plt.plot(df['lifeMale'], color='blue')
plt.ylabel('Life Expectency in Years')
plt.title('Life Expectency for Mele')

plt.subplot(1, 2, 2)
plt.plot(df['lifeFemale'], color='green')
plt.ylabel('Life Expectancy in Years')
plt.title('Life Expectency of Women')
plt.show()

print('The life expectancies show similarity for men and women')
```

```python
plt.scatter(
    x=df['educationMale'],
    y=df['educationFemale'],
    color='green',
    marker='*', s=30
)
plt.ylabel('Female Education')
plt.xlabel('Male Education')
plt.show()
```
