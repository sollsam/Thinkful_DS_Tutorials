```python
import pandas as pd
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
sns.set(style='whitegrid')
```

```python

df = pd.read_csv("Indoor Thermometer_export202012011022.csv")
```

```python
df = df.rename(columns={"Timestamp for sample frequency every 30 min min":'Time', 'Temperature_Fahrenheit': 'Temprature', 'Relative_Humidity': 'Humidity'})
```

```python
df.rename(columns={"Timestamp for sample frequency every 30 min min":"Time"}, inplace = True)#, 'Temperature_Fahrenheit': 'Temprature', 'Relative_Humidity': 'Humidity'}, inplace = True)
```

```python
df.set_axis(['Timestamp', 'Temprature', 'Humidity'], axis=1, inplace=True)
```

```python
df['Timestamp'] = pd.to_datetime(df['Timestamp'])
```

```python
plt.figure(figsize = (16,6))
g = sns.lineplot(data=df, x="Timestamp", y="Temprature")
g.axhline(y=72, linewidth=1, color='g', label = 'Room temprature')
mean = df.Temprature.mean()
plt.axhline(mean, color='r', linestyle='--', label = 'Average temp')
plt.show()
```

Green line represents room temprature which is 72F and the red line shows the average temprature in the house 

```python
plt.figure(figsize = (16,6))
g = sns.lineplot(data=df, x="Timestamp", y="Humidity")
g.axhline(y=60, linewidth=1, color='g', label = 'Max comfort range')
g.axhline(y=30, linewidth=1, color='g', label = 'Min comfort range')
mean = df.Humidity.mean()
plt.axhline(mean, color='r', linestyle='--', label = 'Average temp')
plt.show()
```
