<h2>A/B test</h2>

**The two samples**

The Attrition and Retained groups make up our two samples. 

We will measure the mean job satisfaction for the two groups.

**Hypothesis** 

We expect that people with low job satisfaction level are more likely to leave the company.   

To test the hypothesis we will simply calculate the mean job satisfaction level for each group. 

**Outcome**

If the hypothesis is true, we will conclude that low job satisfaction level increases the rate of attrition. If the Attrition group's job satisfaction level is equal to or greater than the Retained group, we will conclude that low job satisfaction doesn't have an effect on employee attrition. 

<h2>Evaluating A/B Test Result Using T-Test</h2>

Once we calculated the means for each group we have to check whether the difference is a result of actual difference in the groups or just a random chance. This can be done using null hypothesis testing. 

$Hypothesis$: People with low job satisfaction level are more likely to leave the company.

$Null$ $hypothesis$: There is no relation between job satisfaction level and employee attrition. 

We test the null hypothesis by calculating the t-value and the corresponding p-value. For a given t-value, a p-value of less than 0.05 would mean that it is less likely that the difference between the means is because of random chance and hence we would reject the null hypothesis. A p-value of 0.05 or above would mean that it is more likely that the difference in means is due to random chance than actual difference in the population and as a result we wouldn't be able to reject the null hypothesis.

Since we are testing for two possible outcomes, if we reject the null hypothesis, it would mean that we can conclude that our actual hypothesis is true.
