```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline 

df = pd.read_csv('Employee-Attrition.csv')
#df = sns.load_dataset('Employee-Attrition.csv')
df.head()
```

<h2>Choose one variable and plot that variable four different ways.</h2>

```python
sns.distplot(df.Age);
print('The distribution of Age variable with a density curve')
```

```python
print('Histogram with a rug at each observation')
sns.distplot(df.Age, kde=False, rug='True')
plt.show()
```

```python
print('Box plot showing the range of age observation')
sns.boxplot(df.Age)
plt.show()
```

```python
sns.violinplot(x=df.Age)
plt.show()
```

<h2>Choose two continuous variables, and plot them three different ways.</h2>

```python
sns.jointplot(x=df.TotalWorkingYears, y=df.MonthlyIncome)
sns.set_style('whitegrid')
print('Monthly income is positively correlated with experience')
plt.show()
```

```python

sns.jointplot(x=df.TotalWorkingYears, y=df.MonthlyIncome, data=df, kind="kde");
sns.set_style('whitegrid')
print('Most values cluster at the third quadrant, most workers have just less than ten years experience and earn around 5000 per month')
plt.show()
```

```python
sns.jointplot(x=df.TotalWorkingYears, y=df.MonthlyIncome, kind='hex', color='k')
sns.set_style('white')
plt.show()
```
